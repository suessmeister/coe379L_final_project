```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers


x = np.linspace(0, 2*np.pi, 2000).reshape(-1, 1) # sine wave dataset in python!! 
y = np.sin(x).reshape(-1, 1)


x_norm = (x - np.pi) / np.pi # normalize across pi 


model = keras.Sequential([ # tinyNN 
    layers.Input(shape=(1,)),
    layers.Dense(32, activation='tanh'),
    layers.Dense(32, activation='tanh'),
    layers.Dense(1)
])

model.compile(
    optimizer=keras.optimizers.Adam(0.01),
    loss='mse'# optimize for mean squared error and hopefully MAE is ok??
)



model.fit(x_norm, y, epochs=200, batch_size=32, verbose=1)

# should be simple to test 
test_val = np.array([[0.5]])
test_val = (test_val - np.pi) / np.pi   # normalize same way

print("Predicted:", model.predict(test_val))
```

```python
# Evaluate on the full dataset
mse = model.evaluate(x_norm, y, verbose=0)
print("Mean Squared Error (MSE) on training data:", mse)

# get MAE
y_pred = model.predict(x_norm)
mae = np.mean(np.abs(y_pred - y))
print("Mean Absolute Error (MAE) on training data:", mae)


test_val = np.array([[0.5]]) # this line was generated by AI 
test_val_norm = (test_val - np.pi) / np.pi  # normalize same way
pred = model.predict(test_val_norm)
print("Predicted value for input 0.5:", pred[0][0])

model.save("artifacts/sine_model.h5")
```
